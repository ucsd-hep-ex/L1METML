```python
import numpy as np
from utils import convertXY2PtPhi
import matplotlib.pyplot as plt

sample = "TTbar"
mlmet_predictions = np.load(
    f"/home/users/dprimosc/L1METML/experiments/l1metml_v1/25May21_normf100_clip_140X/{sample}_feature_array_MLMET.npy"
)
pumet_predictions = np.load(
    f"/home/users/dprimosc/L1METML/experiments/l1metml_v1/25May21_normf100_clip_140X/{sample}_feature_array_PUMET.npy"
)
mlmet_targets = np.load(
    f"/home/users/dprimosc/L1METML/experiments/l1metml_v1/25May21_normf100_clip_140X/{sample}_target_array_MLMET.npy"
)
pumet_targets = np.load(
    f"/home/users/dprimosc/L1METML/experiments/l1metml_v1/25May21_normf100_clip_140X/{sample}_target_array_PUMET.npy"
)
```

```python
# Plot histograms for each dimension separately, overlapping on same plot

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# MLMET Predictions
axs[0, 0].hist(mlmet_predictions[:, 0], bins=50, color="blue", alpha=0.6, label="Dim 0")
axs[0, 0].hist(mlmet_predictions[:, 1], bins=50, color="cyan", alpha=0.6, label="Dim 1")
axs[0, 0].set_title("MLMET Predictions")
axs[0, 0].set_xlabel("Value")
axs[0, 0].set_ylabel("Frequency")
axs[0, 0].legend()

# PUMET Predictions
axs[0, 1].hist(
    pumet_predictions[:, 0], bins=50, color="green", alpha=0.6, label="Dim 0"
)
axs[0, 1].hist(pumet_predictions[:, 1], bins=50, color="lime", alpha=0.6, label="Dim 1")
axs[0, 1].set_title("PUMET Predictions")
axs[0, 1].set_xlabel("Value")
axs[0, 1].set_ylabel("Frequency")
axs[0, 1].legend()

# MLMET Targets
axs[1, 0].hist(mlmet_targets[:, 0], bins=50, color="red", alpha=0.6, label="Dim 0")
axs[1, 0].hist(mlmet_targets[:, 1], bins=50, color="magenta", alpha=0.6, label="Dim 1")
axs[1, 0].set_title("MLMET Targets")
axs[1, 0].set_xlabel("Value")
axs[1, 0].set_ylabel("Frequency")
axs[1, 0].legend()

# PUMET Targets
axs[1, 1].hist(pumet_targets[:, 0], bins=50, color="orange", alpha=0.6, label="Dim 0")
axs[1, 1].hist(pumet_targets[:, 1], bins=50, color="gold", alpha=0.6, label="Dim 1")
axs[1, 1].set_title("PUMET Targets")
axs[1, 1].set_xlabel("Value")
axs[1, 1].set_ylabel("Frequency")
axs[1, 1].legend()

plt.tight_layout()
plt.show()
```

```python
# Compute magnitude and phi for MLMET and PUMET predictions and targets
mlmet_pred_magphi = convertXY2PtPhi(mlmet_predictions)
pumet_pred_magphi = convertXY2PtPhi(pumet_predictions)
mlmet_targ_magphi = convertXY2PtPhi(mlmet_targets)
pumet_targ_magphi = convertXY2PtPhi(pumet_targets)

# Plot 1: Magnitude distributions
fig1, axs1 = plt.subplots(2, 2, figsize=(12, 10))

# MLMET Predictions - Magnitude
axs1[0, 0].hist(
    mlmet_pred_magphi[:, 0], bins=50, color="blue", alpha=0.7, label="MLMET Predictions"
)
axs1[0, 0].set_title("MLMET Predictions - Magnitude")
axs1[0, 0].set_xlabel("Magnitude")
axs1[0, 0].set_ylabel("Frequency")
axs1[0, 0].legend()

# PUMET Predictions - Magnitude
axs1[0, 1].hist(
    pumet_pred_magphi[:, 0],
    bins=50,
    color="green",
    alpha=0.7,
    label="PUMET Predictions",
)
axs1[0, 1].set_title("PUMET Predictions - Magnitude")
axs1[0, 1].set_xlabel("Magnitude")
axs1[0, 1].set_ylabel("Frequency")
axs1[0, 1].legend()

# MLMET Targets - Magnitude
axs1[1, 0].hist(
    mlmet_targ_magphi[:, 0], bins=50, color="red", alpha=0.7, label="MLMET Targets"
)
axs1[1, 0].set_title("MLMET Targets - Magnitude")
axs1[1, 0].set_xlabel("Magnitude")
axs1[1, 0].set_ylabel("Frequency")
axs1[1, 0].legend()

# PUMET Targets - Magnitude
axs1[1, 1].hist(
    pumet_targ_magphi[:, 0], bins=50, color="orange", alpha=0.7, label="PUMET Targets"
)
axs1[1, 1].set_title("PUMET Targets - Magnitude")
axs1[1, 1].set_xlabel("Magnitude")
axs1[1, 1].set_ylabel("Frequency")
axs1[1, 1].legend()

fig1.suptitle("MET Magnitude Distributions", fontsize=16)
plt.tight_layout()
plt.show()

# Plot 2: Phi distributions
fig2, axs2 = plt.subplots(2, 2, figsize=(12, 10))

# MLMET Predictions - Phi
axs2[0, 0].hist(
    mlmet_pred_magphi[:, 1], bins=50, color="cyan", alpha=0.7, label="MLMET Predictions"
)
axs2[0, 0].set_title("MLMET Predictions - Phi")
axs2[0, 0].set_xlabel("Phi (radians)")
axs2[0, 0].set_ylabel("Frequency")
axs2[0, 0].legend()

# PUMET Predictions - Phi
axs2[0, 1].hist(
    pumet_pred_magphi[:, 1], bins=50, color="lime", alpha=0.7, label="PUMET Predictions"
)
axs2[0, 1].set_title("PUMET Predictions - Phi")
axs2[0, 1].set_xlabel("Phi (radians)")
axs2[0, 1].set_ylabel("Frequency")
axs2[0, 1].legend()

# MLMET Targets - Phi
axs2[1, 0].hist(
    mlmet_targ_magphi[:, 1], bins=50, color="magenta", alpha=0.7, label="MLMET Targets"
)
axs2[1, 0].set_title("MLMET Targets - Phi")
axs2[1, 0].set_xlabel("Phi (radians)")
axs2[1, 0].set_ylabel("Frequency")
axs2[1, 0].legend()

# PUMET Targets - Phi
axs2[1, 1].hist(
    pumet_targ_magphi[:, 1], bins=50, color="gold", alpha=0.7, label="PUMET Targets"
)
axs2[1, 1].set_title("PUMET Targets - Phi")
axs2[1, 1].set_xlabel("Phi (radians)")
axs2[1, 1].set_ylabel("Frequency")
axs2[1, 1].legend()

fig2.suptitle("MET Phi Distributions", fontsize=16)
plt.tight_layout()
plt.show()
```
