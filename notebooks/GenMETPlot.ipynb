```python
import uproot

import matplotlib.pyplot as plt
import numpy as np
```

```python
# Open the ROOT file
file_path = "/ceph/cms/store/user/dprimosc/l1deepmet_data/25Jul8_140X_v0/SingleNeutrino_PU200/FP/140Xv0/perfNano_12509501_136.root"
with uproot.open(file_path) as file:
    tree = file["Events"]
    genMet_pt = tree["genMet_pt"].array(library="np")
    genMet_phi = tree["genMet_phi"].array(library="np")

# Plot genMet_pt
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.hist(genMet_pt, bins=50, color="blue", alpha=0.7)
plt.title("genMet_pt Distribution")
plt.xlabel("genMet_pt")
plt.ylabel("Entries")

# Plot genMet_phi
plt.subplot(1, 2, 2)
plt.hist(genMet_phi, bins=50, color="green", alpha=0.7)
plt.title("genMet_phi Distribution")
plt.xlabel("genMet_phi")
plt.ylabel("Entries")

plt.tight_layout()
plt.show()
```

```python
# Calculate METx and METy from genMet_pt and genMet_phi
met_x = genMet_pt * np.cos(genMet_phi)
met_y = genMet_pt * np.sin(genMet_phi)

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.hist(met_x, bins=50, color="purple", alpha=0.7)
plt.title("METx Distribution")
plt.xlabel("METx")
plt.ylabel("Entries")

plt.subplot(1, 2, 2)
plt.hist(met_y, bins=50, color="orange", alpha=0.7)
plt.title("METy Distribution")
plt.xlabel("METy")
plt.ylabel("Entries")

plt.tight_layout()
plt.show()
```
